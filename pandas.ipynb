```python
import csv
csv_columns = ['Name', 'Age', 'Address']
dict = [
    {'Name':'Aayush','Age':23,'Address':'Katmandu'},
    {'Name':'Shyam','Age':33,'Address':'Patan'}
]

csvfile = open("data.csv", 'w')
writer = csv.DictWriter(csvfile, fieldnames =csv_columns)

writer.writeheader()

writer.writerows(dict)

csvfile.close()
```

```python
import pandas as pd
df1 = pd.DataFrame({'Names':['Ram', 'Shyam', 'Hari'],
                    'Age':[34, 56, 76],
                    'Add':['Kathmandu','Bhaktapur','Lalitpur']})

df2 = pd.DataFrame({'Names':['Ram', 'Shyam', 'Hari'],
                    'Age':[34, 56, 76],
                    'Add':['Kathmandu','Bhaktapur','Lalitpur']})
df = pd.concat([df1,df2]).reset_index()

df
```

```python
import pandas as pd
data =({'Names':['Ram', 'Shyam', 'Hari'],
        'Age':[34, 56, 76],
        'Add':['Kathmandu','Bhaktapur','Lalitpur']})

df1 = pd.DataFrame(data)
df2 = pd.DataFrame(data)
df = pd.concat([df1,df2]).reset_index()

df.to_json('conat_data.json')
df
```
